##### Podemos decir entonces que los años 2017, 2018 y 2019 tienen las siguientes columnas con la siguiente información:

**periodo**: año del dset 

**fecha**: año dia mes

**hora_inicio** hora de inicio de la cuenta de vehiculos (DROP)

**hora_fin** hora de fin de la cuenta de vehiculos (Vamos a usar esta para tomar la hora)

**dia**: dia de la semana 

**estacion**: estaciones de peaje, difieren levemente entre los años (Dellepiane se unifica en 2019, antes habia Dellepiane Liniers y Centro, y se agrega PDB)

**sentido**: Centro o Provincia

**tipo_vehiculo**: Liviano o Pesado

**forma_pago**: Hay varias y varian entre los años.

**cantidad_pasos**: Cantidad de vehiculos

Vamos a estar trabajando con la estación **Illia** y la forma de pago **Efectivo**


```python
#Unimos los 3 datasets, podemos hacerlo asi directamente ya que tienen las mismas columnas
dataset = pd.concat([data_2019, data_2018, data_2017, data_2016, data_2015])

#Nos quedamos solamente con la estación Illia y la forma de pago Efectivo
mask = np.logical_and(dataset.estacion=='Illia', dataset.forma_pago == 'EFECTIVO')
dataset = dataset[mask]

#Quitamos las columnas que no vamos a utilizar
dataset.drop(columns = ['periodo','hora_inicio','forma_pago', 'estacion', 'hora'], inplace = True)

dataset.head()
```

```python
#Pasamos la columna fecha a formato datetime, y le unimos la hora_fin como hora
dataset['fecha2'] = pd.to_datetime(dataset.fecha) + pd.to_timedelta(dataset.hora_fin, unit = 'h')

#Quitamos las columnas fecha y hora_fin ya que las incluimos en el campo fecha2
dataset.drop(columns=['fecha','hora_fin'], inplace = True)

dataset.head()
```

```python
#Corregimos el nombre de fecha2 a fecha
dataset.rename(columns = {'fecha2':'fecha'}, inplace = True) 
dataset.head()
```

```python
#Reordenamos el dataset por fecha
dataset.sort_values('fecha', inplace = True)
dataset.reset_index(drop=True, inplace = True)
dataset
```

```python
#Graficamos nuestros datos, tal como están
plt.figure(figsize = (15,6))
plt.plot(dataset.fecha, dataset.cantidad_pasos)
plt.xlabel('Tiempo')
plt.ylabel('Cantidad de autos, por hora')
plt.show()
```

```python
#Resampleamos nuestros datos para agrupar por dia, y asi tal vez lograr una visualización mas amigable
diario = dataset.resample('D', on = 'fecha').sum()
print(diario.shape)
diario.head()
```

```python
plt.figure(figsize = (15,6))
plt.plot(diario.index, diario.cantidad_pasos)
plt.xlabel('Tiempo')
plt.ylabel('Cantidad de autos, por dia')
plt.show()
```

```python
#tomamos el dataset llamado "diario", que fue actualizado en la primera parte del Proyecto
diario.head()
```

```python
#Como mencionamos anteriormente, tomamos la señal e intentamos buscar las seasonalities de la misma (dataset = diario).
#Buscamos la tendencia que se repite en forma semanal para ver mejor la cantidad de autos que pasan por dia durante una semana.
#Podemos comparar los datos reales con la tendencia y los promedios diarios para ver las variaciones.
#Usamos Seasonal Decompose para realizar esto:
from statsmodels.tsa.seasonal import seasonal_decompose

result = seasonal_decompose(diario, model='additive')
```

```python
import matplotlib
matplotlib.rcParams['figure.figsize'] = [12,12]
result.plot()
plt.show()
```

```python
#Resultados encontrados
result.seasonal[2:9]
```

```python
#Graficamos los resultados
plt.figure(figsize = (15,6))
plt.plot(result.seasonal[:7].values)
plt.xlabel('Dia de la semana')
plt.ylabel('Cantidad de autos: Variación Promedio semanal vs Dia de la Semana')
plt.show()
```

```python
#Definimos los dias de la semana a graficar
week = ['Viernes', 'Sabado', 'Domingo', 'Lunes', 'Martes', 'Miercoles', 'Jueves']
```

```python
#Realizamos el mismo gráfico, pero incluyendo los días de la semana.
plt.figure(figsize = (15,6))
plt.xticks(range(len(week)), week, rotation='horizontal')
plt.plot(result.seasonal[:7].values)
plt.xlabel('Dia de la semana')
plt.ylabel('Cantidad de autos: Variación Promedio semanal vs Dia de la Semana')
plt.show()
```

```python
#Podemos mostrar un mismo grafico en donde incluimos las 3 variables
aproximacion = result.seasonal + result.trend
```

```python
plt.figure(figsize = (15,6))
plt.title('Serie Original y Apoximacion')
plt.plot(diario.index, diario.values)
plt.plot(aproximacion.index, aproximacion.values)
plt.plot(diario.index, result.trend)
plt.xlim([datetime.date(2019, 1, 1), datetime.date(2019, 12, 31)])
plt.show()
```

```python
plt.figure(figsize = (16,8))
plt.hist(result.resid, bins = 30)
plt.xlabel('Error en Cantidad de autos')
plt.ylabel('Cuentas = cuantas veces se repite el error por cantidad de autos')
plt.show()
```

```python
from statsmodels.tsa.stattools import acf
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import datetime
sns.set()
```

```python
#Quitamos los Outliers para reducir errores
diario.cantidad_pasos[diario.cantidad_pasos < 20000] = np.nan
diario1 = diario.interpolate()
diario1.shape
```

```python
senial = diario1.cantidad_pasos
senial.head()
```

```python
look_back = int(365*0.2) # usamos 1/5 de año de datos para predecir el valor siguiente
```

```python
# se arma la matriz vacia para luego ir llenandola. 
N = senial.shape[0]
X = np.zeros((N - look_back,look_back))
y = np.zeros(N - look_back)
print(X.shape, y.shape)
```

```python
for i in range(X.shape[0]):
    X[i,:] = senial.iloc[i:i+look_back]
    y[i] = senial.iloc[i+look_back]
```

```python
X.shape
```

```python
y.shape
```

```python
N_train = 800 # cantidad de instancias que vamos a usar para entrenar. 
N_test = 204 #cantidad de puntos que vamos a usar para evaluar
N_total = N_train + N_test

length_total = N_train + N_test + look_back

X_train = X[:N_train,:]
y_train = y[:N_train]
print(X_train.shape, y_train.shape)

X_test = X[N_train:N_train+N_test,:]
y_test = y[N_train:N_train+N_test]
print(X_test.shape, y_test.shape)
```

```python
from sklearn.metrics import mean_squared_error
from sklearn.ensemble import RandomForestRegressor

rf = RandomForestRegressor(n_estimators=400, random_state=42,max_depth=4 ,n_jobs = 4)
rf.fit(X_train,y_train)
```

```python
#Importancia de los parametros de los días anteriores, importancia de los features por dia para calcular los días futuros.
plt.figure(figsize = (12,7))
plt.scatter(np.arange(rf.feature_importances_.size),rf.feature_importances_)
# for x in np.arange(3)*52:
# plt.axvline(x, linestyle ='--', color= 'r')
plt.show()
```

```python
y_pred = rf.predict(X)
y_train_pred = rf.predict(X_train)
y_test_pred = rf.predict(X_test)
```

```python
MSE = mean_squared_error(y_train, y_train_pred)
print('Error en Train:',np.sqrt(MSE))

MSE = mean_squared_error(y_test, y_test_pred)
print('Error en Test:',np.sqrt(MSE))
```

```python
plt.figure(figsize=(16,8))

plt.plot(senial.index[:length_total], senial[:length_total],lw = 0.75, label = 'Serie Original')
plt.plot(senial.index[:N_train + look_back], senial[:N_train + look_back],'--', label = 'Usado para entrenar')

plt.plot(senial.index[look_back: look_back + N_train], y_train_pred,'-.',label = 'Predicho en Train')
plt.plot(senial.index[N_train+look_back: length_total], y_test_pred,'-.', lw = 4, label = 'Predicho en Test')
plt.legend()
plt.xlabel('Fechas')
plt.ylabel('Cantidad de autos')
plt.show()
```

```python
y_test_pred_benchmark = y[-1 + N_train:N_train+N_test - 1]
```

```python
plt.figure(figsize=(12,4))
plt.plot(senial.index[:length_total], senial[:length_total],lw = 0.75, label = 'Serie Original')
plt.plot(senial.index[N_train+look_back: length_total], y_test_pred_benchmark,'-.', lw = 2, label = 'Predicho en Test, Benchmark')
plt.xlim(senial.index[N_train+look_back], senial.index[length_total])
plt.xlabel('Fechas')
plt.ylabel('Cantidad de Autos')
plt.legend()
plt.show()
```

```python
#Verificamos el Error del Benchmark vs El modelo aplicado anteriormente
MSE = mean_squared_error(y_test, y[-1 + N_train:N_train+N_test - 1])
print(np.sqrt(MSE))
```

```python
y_ms_test_pred = []
x = X_test[0,:] # arrancamos con la primera instancias del conjunto de test
for i in range(y_test.size):
    y_os_pred = rf.predict(x.reshape(1,-1))
    y_ms_test_pred.append(y_os_pred)
    x = np.append(x[1:], y_os_pred)
y_ms_test_pred = np.array(y_ms_test_pred)
```

```python
plt.figure(figsize=(16,8))
plt.plot(senial.index[:length_total], senial[:length_total],lw = 0.75, label = 'Serie Original')
plt.plot(senial.index[:N_train + look_back], senial[:N_train + look_back],'--', label = 'Usado para entrenar')
plt.plot(senial.index[look_back: look_back + N_train], y_train_pred,'-.',label = 'Predicho en Train')
plt.plot(senial.index[N_train+look_back: length_total], y_ms_test_pred,'-.', lw = 4, label = 'Predicho en Test. Multi-Step')
plt.legend()
plt.xlabel('Fechas')
plt.ylabel('Cantidad de Autos')
plt.show()
```

```python
MSE = mean_squared_error(y_test, y_ms_test_pred)
print('Error en Test:',np.sqrt(MSE))
```
